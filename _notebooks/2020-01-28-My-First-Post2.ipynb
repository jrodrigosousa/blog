```python
from fastbook import *
from fastai.vision.widgets import *
```

```python
path = Path("rostos")
learner = load_learner("../input/modelo-treinado-homem-mulher/homem_mulher.pkl")
```

```python
btn_upload = widgets.FileUpload()
btn_upload
```

```python
resultado = widgets.Label()
```

resultado.value = f"Previsão: {pred}, Probabilidade {probs[pred_idx]}"

```python
classificar = widgets.Button(description="Classificar")
classificar
```

```python
def on_click_classificar(change):
    imagem = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl:
        display(imagem.to_thumb(128, 128))
    pred, pred_idx, probs = learner.predict(imagem)
    resultado.value = f"Previsão: {pred}, Probabilidade {probs[pred_idx]}"
    
classificar.on_click(on_click_classificar)
```

```python
VBox([widgets.Label('Selecione uma foto'), btn_upload, classificar, out_pl, resultado])
```
